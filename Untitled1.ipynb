# Filtering and ordering

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Personal\Desktop\programación\world_population.csv')
df
```

```python
df[df['Rank'] <= 10]
```

```python
specific_countries = ['Mexico','Brazil']
df[df['Country'] .isin(specific_countries)]
```

```python
df[df['Country'] .str.contains('United')]
```

```python
df2 = df.set_index('Country')
df2
```

```python
df2.filter(items =['Continent', 'CCA3'])
```

```python
df2.filter(items =['Zimbabwe'], axis = 0)
```

```python
df2.loc['Yemen']
```

```python
df2.iloc[4]
```

```python
df2[df2['Rank'] <= 10].sort_values(by='Rank', ascending=True)
```

```python
df2[df2['Rank'] <= 10].sort_values(by='Rank', ascending=False)
```

```python
df2[df2['Rank'] <= 10].sort_values(by=['Continent','Country'], ascending=[False,True])
```

```python
df2.filter(like = 'United', axis =0)
```
